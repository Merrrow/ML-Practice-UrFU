#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(s):
    """Возвращает количество гласных (a, e, i, o, u) в строке. Регистр не учитывается."""
    return sum(1 for ch in s.lower() if ch in 'aeiou')

def tests():
    assert count_vowels("xyz") == 0
    assert count_vowels("") == 0
    assert count_vowels("aeiou") == 5
    assert count_vowels("AEIOU") == 5
    assert count_vowels("AaEeIiOoUu") == 10
    assert count_vowels("rhythm") == 0
    assert count_vowels("12345") == 0
    print("All tests passed!")

tests()
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def all_unique(s):
    """Возвращает True, если все символы в строке уникальны, иначе False."""
    return len(set(s)) == len(s)

def tests():
    assert all_unique("abcde") == True
    assert all_unique("hello") == False
    assert all_unique("") == True
    assert all_unique("a") == True
    assert all_unique("aa") == False
    assert all_unique("12345") == True
    assert all_unique("12341") == False
    assert all_unique("a b c") == False  # пробел повторяется
    assert all_unique("aAbBcC") == True  # регистр разный
    assert all_unique("!@#$%") == True
    print("All tests passed!")

tests()
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_set_bits(n):
    """Возвращает количество единичных бит в положительном целом числе."""
    return bin(n).count('1')

def tests():
    assert count_set_bits(0) == 0
    assert count_set_bits(1) == 1
    assert count_set_bits(2) == 1
    assert count_set_bits(3) == 2
    assert count_set_bits(7) == 3
    assert count_set_bits(8) == 1
    assert count_set_bits(15) == 4
    assert count_set_bits(16) == 1
    assert count_set_bits(255) == 8
    assert count_set_bits(1023) == 10
    print("All tests passed!")

tests()
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    """Возвращает количество шагов умножения цифр до получения одной цифры."""
    steps = 0
    while n >= 10:
        prod = 1
        while n > 0:
            prod *= n % 10
            n //= 10
        n = prod
        steps += 1
    return steps

def tests():
    assert magic(39) == 3
    assert magic(4) == 0
    assert magic(999) == 4
    assert magic(0) == 0
    assert magic(10) == 1
    assert magic(25) == 2
    assert magic(77) == 4
    assert magic(888) == 3
    assert magic(1111) == 1
    assert magic(277777788888899) == 11
    print("All tests passed!")

tests()
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
import numpy as np

def mse(pred, true):
    """Среднеквадратическая ошибка (MSE) двух векторов одинаковой длины."""
    pred = np.asarray(pred, dtype=float)
    true = np.asarray(true, dtype=float)
    return np.mean((pred - true) ** 2)

def tests():
    import numpy as np
    assert mse([1, 2, 3], [1, 2, 3]) == 0.0
    assert mse([0, 0, 0], [1, 1, 1]) == 1.0
    assert mse([1, 1], [2, 3]) == (1**2 + 2**2) / 2  # 2.5
    assert abs(mse([1.5, 2.5], [1, 2]) - 0.25) < 1e-9
    assert mse([-1, -2], [1, 2]) == 10.0
    # целочисленные векторы одинаковой длины
    assert mse([0], [0]) == 0.0
    print("All tests passed!")

tests()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def factorize(n):
    """Разложение на простые множители: возвращает строку вида (n1**p1)(n2)..."""
    if n <= 0:
        raise ValueError("Positive integer required")
    res = []
    d = 2
    while d * d <= n:
        if n % d == 0:
            cnt = 0
            while n % d == 0:
                n //= d
                cnt += 1
            if cnt == 1:
                res.append(f"({d})")
            else:
                res.append(f"({d}**{cnt})")
        d += 1
    if n > 1:
        res.append(f"({n})")
    return ''.join(res)

def tests():
    assert factorize(1) == ""
    assert factorize(2) == "(2)"
    assert factorize(3) == "(3)"
    assert factorize(4) == "(2**2)"
    assert factorize(6) == "(2)(3)"
    assert factorize(8) == "(2**3)"
    assert factorize(9) == "(3**2)"
    assert factorize(86240) == "(2**5)(5)(7**2)(11)"
    assert factorize(13) == "(13)"
    assert factorize(100) == "(2**2)(5**2)"
    assert factorize(999) == "(3**3)(37)"
    print("All tests passed!")

tests()
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid_height(n):
    """Возвращает k, если n = 1^2 + 2^2 + ... + k^2, иначе 'It is impossible'."""
    if n <= 0:
        return "It is impossible"
    total = 0
    k = 1
    while True:
        total += k * k
        if total == n:
            return k
        if total > n:
            return "It is impossible"
        k += 1

def tests():
    assert pyramid_height(1) == 1
    assert pyramid_height(5) == 2
    assert pyramid_height(14) == 3
    assert pyramid_height(30) == 4
    assert pyramid_height(55) == 5
    assert pyramid_height(91) == 6
    assert pyramid_height(0) == "It is impossible"
    assert pyramid_height(-5) == "It is impossible"
    assert pyramid_height(2) == "It is impossible"
    assert pyramid_height(4) == "It is impossible"
    assert pyramid_height(6) == "It is impossible"
    assert pyramid_height(13) == "It is impossible"
    assert pyramid_height(15) == "It is impossible"
    print("All tests passed!")

tests()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(n):
    """Проверяет, сбалансировано ли число: сумма цифр до средних равна сумме после.
    Средние цифры (одна или две) исключаются из сумм."""
    s = str(n)
    L = len(s)
    if L % 2 == 1:
        mid = L // 2
        left = s[:mid]
        right = s[mid + 1:]
    else:
        # две средние цифры: индексы (L//2 - 1) и (L//2)
        left = s[:L // 2 - 1]
        right = s[L // 2 + 1:]
    return sum(map(int, left)) == sum(map(int, right))

def tests():
    assert is_balanced(1221) == True
    assert is_balanced(12321) == True
    assert is_balanced(1) == True       # нет цифр слева/справа -> суммы 0
    assert is_balanced(12) == True      # две средние, пустые суммы
    assert is_balanced(1001) == True
    assert is_balanced(10001) == True
    assert is_balanced(1111) == True
    assert is_balanced(11111) == True

    assert is_balanced(1234) == False   # 1 vs 4
    assert is_balanced(1230) == False   # 1 vs 0
    assert is_balanced(123) == False    # 1 vs 3
    assert is_balanced(12345) == False  # 1+2=3 vs 4+5=9
    print("All tests passed!")

if __name__ == '__main__':
    tests()
```
